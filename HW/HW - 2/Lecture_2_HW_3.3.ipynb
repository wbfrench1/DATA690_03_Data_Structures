### Finger Exercise 3.3.1 

What would the code in Figure 3.4 do if the statement x = 25 were replaced by x = -25.

```python
def Bisection_sqrt (f_square: float, f_epsilon) -> int:
  i_step_count = 0 
  f_low = 0.0
  f_high = max(1.0, f_square)
  f_ans = ( f_high + f_low)/2.0
  while abs(f_ans**2 - f_square) >= f_epsilon:
    print('low =', f_low, 'high =', f_high, 'ans =', f_ans)
    i_step_count += 1
    if f_ans**2 < f_square:
      f_low = f_ans
    else:
      f_high = f_ans
    f_ans = (f_high + f_low)/2.0
  print('In', i_step_count, 'steps Bisection found that the square root of', f_square, 'is about', f_ans)
  return i_step_count
```

The result of running Bisection_sqrt(-25,.01) is an infinite loop because  f_low starts at 0 and f_high approaches 0 and so the while loop test is always greater than epsilon.

What would have to be changed to make the code in Figure 3.4 work for finding an approximation to the cube root of both negative and positive numbers?

```python
def Bisection_cube (f_cube: float, f_epsilon) -> int:
  i_step_count = 0
  if f_cube > 0: 
    f_low = 0.0
    f_high = max(1.0, f_cube)
  else:
    f_low = min(f_cube, -1.0)
    f_high = 0.0
  f_ans = ( f_high + f_low)/2.0
  while abs(f_ans**3 - f_cube) >= f_epsilon:
    i_step_count += 1
    if f_ans**3 < f_cube:
      f_low = f_ans
    else:
      f_high = f_ans
    f_ans = (f_high + f_low)/2.0
  print('In', i_step_count, 'steps Bisection found that the square root of', f_cube, 'is about', f_ans)
  return i_step_count
```
