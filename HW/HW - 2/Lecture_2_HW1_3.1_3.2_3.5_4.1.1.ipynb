### Finger Exercise 3.1

<p>Write a program that asks the user to enter an integer and prints two integers, root and pwr, such that 1 < pwr < 6 and root**pwr is equal to the integer entered by the user.  If no such pair of integers exists, it should print a message to that effect.

```python
# Ask the user for an interger and cast the entry to type integer
i_usr_num = int(input('Enter an integer:'))

# Test for the sign of user input to determine proper sign of root and step 
if i_usr_num > 0:
  i_root = 1
  i_step = 1
else:
  i_root = -1
  i_step = -1

# initialize the power to 1
i_pwr = 1

# loop while power is less than 6 and root^power != number
## Note that abs is not used here to account for proper sign
while (i_pwr < 6) and (i_root ** i_pwr != i_usr_num):
  # initialize root to 1
  i_root = i_step
  # increment power to 2 s.t. power > 1
  i_pwr += 1
  # use absolute value here to account for sign of user integer
  while (abs(i_root) ** i_pwr < abs(i_usr_num)):
    # step the root according to the sign
    i_root = i_root + i_step
    
if i_root ** i_pwr == i_usr_num:
  print(i_root,' raised to the ', i_pwr, 'power = ', i_usr_num)
else:
  print('There are no integers root and 1 < pwr < 6 such that root^pwr = ', i_usr_num)


```

### Finger Exercise 3.2

Let s be a string that contains a sequence of decimal numbers separated by commas, e.g., s = '1.23,2.4,3.123'.  Write a program that prints the sum of the numbers in s.

```python
# define s to be the string containing decimals separated by commans
s = '1.23,2.4,3.123'
# initialize the total to 0
f_total = 0

# loop through the sequence of strings generated by splitting s at the commas
for num in s.split(','):
  # convert the string num to a float and sum it with f_total
  f_total += float(num)

print('The sum of the decimals separated by commans in ',s, 'is: ', f_total)
```

### Finger Exercise 3.5

Add some code to the implementation of Netwon-Raphson that keeps track of the number of iterations used to find the root.  Use that code as part of a program that compares the efficiency of Newton-Raphson and bisection search.


```python
def NewtonRaphson_sqrt(f_square: float, f_epsilon) -> int:
  f_guess = f_square/2.0
  i_step_count = 0
  while abs(f_guess * f_guess - f_square) >= f_epsilon:
    f_guess = f_guess - ((f_guess**2) - f_square)/(2 * f_guess)
    i_step_count += 1
  print('In', i_step_count, 'steps NR found that the square root of', f_square, 'is about', f_guess)
  return i_step_count
```

```python
def Bisection_sqrt (f_square: float, f_epsilon) -> int:
  i_step_count = 0 
  f_low = 0.0
  f_high = max(1.0, f_square)
  f_ans = ( f_high + f_low)/2.0
  while abs(f_ans**2 - f_square) >= f_epsilon:
    i_step_count += 1
    if f_ans**2 < f_square:
      f_low = f_ans
    else:
      f_high = f_ans
    f_ans = (f_high + f_low)/2.0
  print('In', i_step_count, 'steps Bisection found that the square root of', f_square, 'is about', f_ans)
  return i_step_count
```

```python
f_epsilon = 0.01
f_square = float(input('Enter a positive number to find its square root within ' + str(f_epsilon) +' accuracy: '))

i_bi_steps = Bisection_sqrt(f_square,f_epsilon)
i_nr_steps = NewtonRaphson_sqrt(f_square,f_epsilon)

if (i_bi_steps < i_nr_steps):
  print('Bisection found the answer in', i_bi_steps, 'which is', \
        i_nr_steps - i_bi_steps, 'fewer steps than Newton-Raphson' )
elif (i_bi_steps > i_nr_steps):
  print('Newton-Raphson found the answer in', i_nr_steps, 'which is', \
        i_bi_steps - i_nr_steps , 'fewer steps than Bisection' )
else:
  print('Both Newton-Raphson and Bisection found the answer in ', i_bi_steps, 'steps.')
```

### Finger Exercise 4.1.1

Write a function isIn that accepts two strings as arguments and returns True if either string occurs anywhere in the other, and False otherwise.

```python
def isIn (str_1: str, str_2: str) -> bool:
  '''This function assumes it receives two strings, str1 and str2 and it
  returns True if either str occurs in the other.  Note capitalization matters'''

  if (str_1 in str_2) or (str_2 in str_1):
    return True
  
  return False
```
